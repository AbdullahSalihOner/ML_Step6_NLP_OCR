```python
import keras_ocr
import matplotlib.pyplot as plt
import numpy as np  
```

```python
pipeline = keras_ocr.pipeline.Pipeline() # keras_ocr otoamtlı olarak modeli indirir ve kullanır / keras_ocr automatically downloads and uses the model
```

```python
images = [keras_ocr.tools.read(img) for img in ['resimler/test.jpg']] # test.jpg dosyasını okur / reads test.jpg file
```

```python
images
```

```python
predictions = pipeline.recognize(images) # resimdeki yazıları tahmin eder / predicts the text in the image
```

```python
#Resim ve tahminleri görselleştirir / Visualizes the image and predictions
fig , ax = plt.subplots(figsize=(18,18)) # tek bir resim için 18x18 boyutunda bir figür oluşturur tek eksende / creates a figure with a size of 18x18 for a single image on a single axis
for image, prediction in zip(images, predictions):
    keras_ocr.tools.drawAnnotations(image=image, predictions=prediction, ax=ax) # resim ve tahminleri çizer / draws the image and predictions
```

```python
predicted_image = predictions[0] # tahmin edilen resim   / predicted image
for text,box in predicted_image:
    print(text) # tahmin edilen yazıları yazdırır / prints the predicted texts
```

```python
import keras_ocr
import cv2 # opencv kütüphanesi / opencv library
import matplotlib.pyplot as plt
```

```python
#Modelleri yükleme / Loading the models
pipeline = keras_ocr.pipeline.Pipeline() # keras_ocr otoamtlı olarak modeli indirir ve kullanır / keras_ocr automatically downloads and uses the model
```

```python
original_image = keras_ocr.tools.read('resimler/resim3.jpg') # resim.jpg dosyasını okur / reads resim.jpg file
```

```python
#Resmi bulanıklaştırma / Blurring the image
blurred_image = cv2.GaussianBlur(original_image, (15,15), 0) # resmi bulanıklaştırır / blurs the image
```

```python
#Ocr işlemi / Ocr process
prediction = pipeline.recognize([blurred_image]) # resimdeki yazıları tahmin eder / predicts the text in the image
```

```python
fig, ax = plt.subplots(figsize=(18,18)) # tek bir resim için 20x20 boyutunda bir figür oluşturur tek eksende / creates a figure with a size of 20x20 for a single image on a single axis
keras_ocr.tools.drawAnnotations(image=blurred_image, predictions=prediction[0], ax=ax) # resim ve tahminleri çizer / draws the image and predictions
```

```python
predicted_image = prediction[0] # tahmin edilen resim  / predicted image
for text,box in predicted_image:
    print(text) # tahmin edilen yazıları yazdırır / prints the predicted texts
```
